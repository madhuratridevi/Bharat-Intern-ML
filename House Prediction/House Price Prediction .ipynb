```python
import pandas as pd
from sklearn.linear_model import LinearRegression 
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('C:\ML Internship\Bharat_Intern\House Prediction\dataset.csv')
```

```python
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(dataset.area,dataset.price,color='blue',marker='*')
```

```python
X = dataset.drop('price',axis='columns')
X

Y = dataset.price
Y
```

```python
model = LinearRegression()
model.fit(X,Y)
```

```python
x=int(input("Enter the Square feet of the land : "))
LandAreainSqFt=[[x]]
PredictedmodelResult = model.predict(LandAreainSqFt)
print(PredictedmodelResult)
```

```python
m=model.coef_
print(m)
```

```python
b=model.intercept_
print(b)
```

```python
y = m*x + b
print("The Price of {0} Square feet Land is: {1}".format(x,y[0]))
```
